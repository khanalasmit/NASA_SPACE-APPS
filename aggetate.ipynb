```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
kepler= pd.read_csv("E:\\Space  app\\data\\kepler.csv")
tess = pd.read_csv("E:\\Space  app\\data\\tess.csv")
```

```python
pd.set_option('display.max_columns', None)
```

```python
kepler.head()
```

```python
kepler = kepler[kepler['koi_disposition'] != 'CANDIDATE']

```

```python
kepler.shape
```

```python
tess['tfopwg_disp'].value_counts()
```

```python
tess = tess[~tess['tfopwg_disp'].isin(['PC', 'APC'])]
```

```python
tess.shape
```

```python
label_map_tess = {
    'CP' : 1,
    'KP' : 1,
    'FP' : 0,
    'FA' : 0
}
```

```python
label_map_kepler = {
    "CONFIRMED": 1,
    
    "FALSE POSITIVE": 0
}
```

```python
kepler['label'] = kepler['koi_disposition'].map(label_map_kepler)
tess['label'] = tess['tfopwg_disp'].map(label_map_tess)
```

```python
kepler.head()
```

```python
tess.head()
```

```python
koi_features = kepler[['koi_period','koi_duration','koi_depth','koi_prad','koi_insol',
                       'koi_teq','koi_steff','koi_srad','koi_slogg','koi_kepmag']].copy()
koi_features.columns = ['period','duration','depth','radius','insol','teq','teff','srad','logg','mag']
koi_features['label'] = kepler['label']
koi_features['PlanetId'] = kepler['kepoi_name']
koi_features['StarID']= kepler['kepid']
```

```python
toi_features = tess[['pl_orbper','pl_trandurh','pl_trandep','pl_rade','pl_insol',
                       'pl_eqt','st_teff','st_rad','st_logg','st_tmag']].copy()
toi_features.columns = ['period','duration','depth','radius','insol','teq','teff','srad','logg','mag']
toi_features['label'] = tess['label']
toi_features['planetTd'] = tess['toi']
toi_features['StarID']= tess['tid']
```

```python
toi_features.sample()
```

```python
koi_features.dropna(inplace=True)
toi_features.dropna(inplace=True)
```

```python
koi_features.shape
```

```python
toi_features.shape
```

```python
toi_features.to_csv("E:\\Space  app\\data\\tess_features.csv", index=False)
koi_features.to_csv("E:\\Space  app\\data\\kepler_features.csv", index=False)
```

```python
combined_df = pd.concat([koi_features, toi_features], ignore_index=True)
```

```python
combined_df = combined_df.dropna()
```

```python
scaler = StandardScaler()
X = combined_df[['period','duration','depth','radius','insol','teq','teff','srad','logg','mag']]
X_scaled = scaler.fit_transform(X)
```

```python
final_df = pd.DataFrame(X_scaled, columns=X.columns)
final_df['label'] = combined_df['label'].values
final_df.to_csv('koi_toi_combined_features.csv', index=False)
print(" Combined features CSV ready for ML training.")
```
